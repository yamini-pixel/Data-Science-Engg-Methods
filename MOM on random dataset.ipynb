#### Problem Statement:

Find a simple dataset with a one-humped histogram and model it by matching with the pdfs we learned today. Find the most likely parameters for your pdf using MOM. Use the model to do a prediction.

#### Solution: 

We have generated a random dataset of 1000 numbers with mean = 10 and SD = 2, and have performed the required functions on it.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(12)
data = np.random.gamma(10, 2, 1000)
print(data)
```

```python
# Plot histogram
plt.hist(data, bins=20, density=True)
plt.xlabel("x")
plt.ylabel("Probability density")
plt.show()
```

```python
# Calculate sample mean and variance
mean = np.mean(data)
var = np.var(data, ddof=1)

# Calculate gamma pdf parameters
alpha = (mean**2) / var
beta = var / mean
print("alpha =", alpha)
print("beta =", beta)
```

```python
from scipy.stats.distributions import gamma
plt.hist(data, bins=20, density=True)
x = np.linspace(3, 45)
plt.plot(x, gamma.pdf(x, alpha, scale=beta))
```

```python
# Calculate gamma pdf for x=11
x = 11
prob = gamma.pdf(x, alpha, scale=1/beta)
print("Probability density for x=11:", prob)
```
